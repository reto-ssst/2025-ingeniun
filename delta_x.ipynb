## Importazione delle librerie necessarie

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generazione del vettore degli argomenti

```python
x=np.linspace(-2,2,201)
# print(x)
```

## Definizione della funzione $f(x)$

Qui definisco la funzione da utilizzare

```python
def f(x):
    return(np.sin(x))
```

## Definizione della pendenza $m(x)$

Derivazione numerica di $f(x)$

```python
def m(x,dx):
    return (f(x+dx)-f(x))/(dx)
```

```python
plt.plot(x, f(x), label='f(x)')
for dx in [2, 1.5, 1, 0.5, 0.1, 1e-9]:
    plt.plot(x, m(x,dx), label=f'm(x, dx={dx})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) e m(x) per x da -10 a 10')
plt.grid(True)
plt.ylim(-2, 5)
plt.legend()
plt.show()
```

```python
dx=np.linspace(0.1,1e-9,201)

plt.plot(dx, m(0,dx), label='m(0,dx)')
plt.xlabel('dx')
plt.ylabel('m(0,dx)')
plt.grid(True)
#plt.ylim(-2, 5)
plt.legend()
plt.show()
```
