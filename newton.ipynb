## Importazione delle librerie necessarie

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definizione della funzione $f(x)$

Qui definisco la funzione da utilizzare

```python
def f(x):
    return(2*x**3+3*x**2-4*x-2)
```

# Funzione pendenza $m(x)$

```python
def m(x):
    dx=1e-5
    return (f(x+dx)-f(x))/(dx)
```

## Grafico di $f(x)$

```python
x=np.linspace(-5,5,201)

plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico di y = f(x) per x da -10 a 10')
plt.grid(True)
plt.ylim(-10, 10)
plt.show()
```

## Funzione per determinare uno zero

Determina uno zero partendo da un valore initziale $x$

```python
def zero(x):
    # x valore iniziale
    while abs(f(x))>1e-6:
        x=x-f(x)/m(x)   # prossima approssimazione
    print('ho trovato uno zero per x =',x)
    #return x
```

```python
zero(-2)
zero(-0.5)
zero(1)
```
