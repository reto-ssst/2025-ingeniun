# Pendenza di una curva

Script python che ci permette di determinare la pendenza della curva di una funzione in qualsiasi punto.

## Importazione delle librerie necessarie

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definizione della funzione $f(x)$

Qui definisco la funzione da utilizzare

```python
def f(x):
    return(x**3+x**2-5*x+2)
```

```python
f(0)
```

Genero il vettore degli argomrenti $x$

```python
x=np.linspace(-10,10,201)
print(x)
```

Genero il vettore delle immagini $y$

```python
y=f(x)
print(y)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico di y = f(x) per x da -10 a 10')
plt.grid(True)
plt.ylim(-10, 10)
plt.show()
```

# Funzione pendenza $m(x)$

```python
def m(x):
    dx=1e-5
    return (f(x+dx)-f(x))/(dx)
```

```python
plt.plot(x, y, label='f(x)')
plt.plot(x, m(x), color='red', label='m(x) (pendenza)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) e m(x) per x da -10 a 10')
plt.grid(True)
plt.ylim(-10, 10)
plt.legend()
plt.show()
```

```python
coeff_m = np.polyfit(x, m(x), 2)
coeff_m_rounded = np.round(coeff_m, 4)
print("Coefficienti della curva m(x) di secondo grado arrotondati a 4 cifre:", coeff_m_rounded)
```
